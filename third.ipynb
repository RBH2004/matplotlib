Scatter plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
colors=[10,49,30,29,126,20,30]
sizes=[500,200,400,600,800,900,400]
plt.scatter(x,y,c=colors,s=sizes,marker='*',edgecolor='g',linewidths=1.5,cmap='BrBG')#also  color can be used in the place of c here
plt.colorbar()
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.title('scatter plot',fontsize=5)
plt.show()

```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
colors=[10,49,30,29,126,20,30]
sizes=[500,200,400,600,800,900,400]
plt.scatter(x,y,c=colors,s=sizes,marker='*',alpha=0.2,edgecolor='g',linewidths=1.5,cmap='BrBG')#also  color can be used in the place of c here
t=plt.colorbar()
t.set_label('colorbar',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.title('scatter plot',fontsize=5)
plt.show()
```
