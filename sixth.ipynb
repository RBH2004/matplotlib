#stemp plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,linefmt=':')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,markerfmt='r+')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,markerfmt='+')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,bottom=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,basefmt='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,label='rafid')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50]
y=[2,2,5,8,3]
plt.stem(x,y,orientation='horizontal')
plt.show()
```
