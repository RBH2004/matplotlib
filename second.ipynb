```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[75,70,60,82]
plt.xlabel('language',fontsize=50)
plt.ylabel('number',fontsize=50)
plt.title('rafid',fontsize=50)
plt.bar(x,y)
plt.show()
```

```python
x=['python','c','c++','java']
y=[75,70,60,82]
plt.xlabel('language',fontsize=10)
plt.ylabel('number',fontsize=10)
plt.title('rafid',fontsize=10)
plt.bar(x,y,width=0.4,color='g')#using width and color parameter
plt.show()
```

```python
x=['python','c','c++','java']
y=[75,70,60,82]
plt.xlabel('language',fontsize=10)
plt.ylabel('number',fontsize=10)
plt.title('rafid',fontsize=10)
c=['b','g','y','r']
plt.bar(x,y,width=0.4,color=c,align='center')#using width and color parameter
plt.show()
```

```python
x=['python','c','c++','java']
y=[75,70,60,82]
plt.xlabel('language',fontsize=10)
plt.ylabel('number',fontsize=10)
plt.title('rafid',fontsize=10)
c=['b','g','y','r']
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=5,linestyle=':')
plt.show()
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=5,linestyle=':',alpha=0.5)#using width and color parameter
plt.show()
```

```python
x=['python','c','c++','java']
y=[75,70,60,82]
z=[20,30,40,50]
plt.xlabel('language',fontsize=10)
plt.ylabel('number',fontsize=10)
plt.title('rafid',fontsize=10)
plt.bar(x,y,width=0.4,color='r',label='graph')#using width and color parameter
plt.bar(x,z,width=0.4,color='y',label='graph2')#using width and color parameter
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','java']
y=[75,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('language',fontsize=15)
plt.ylabel('number',fontsize=15)
plt.title('rafid',fontsize=15)
plt.bar(p,y,width=0.4,color='r',label='graph')#using width and color parameter
plt.bar(p1,z,width=0.4,color='y',label='graph2')#using width and color parameter

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[75,70,60,82]
z=[20,30,40,50]
plt.xlabel('language',fontsize=10)
plt.ylabel('number',fontsize=10)
plt.title('rafid',fontsize=10)

plt.barh(x,y,color='r',label='graph')#using width and color parameter
plt.barh(x,z,color='y',label='graph2')#using width and color parameter

plt.legend()
plt.show()
```
