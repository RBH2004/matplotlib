```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.legend()

plt.show( )
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c)
plt.legend()

plt.show( )
```
